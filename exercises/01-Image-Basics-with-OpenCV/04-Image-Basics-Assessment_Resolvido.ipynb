#### TASK: Draw an empty RED rectangle around the dogs face and display the image in the notebook.

````python
img_fix = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_fix)
````

````python
cv2.rectangle(img = img_fix, pt1 = (200,375), pt2 = (600,725), color=(255,0,0), thickness=5)
plt.imshow(img_fix)
````

#### TASK: Draw a BLUE TRIANGLE in the middle of the image. The size and angle is up to you, but it should be a triangle (three sides) in any orientation.

````python
pt1 = (467,500)
pt2 = (267,900)
pt3 = (667,900)

cv2.line(img=img_fix, pt1=pt1, pt2=pt2, color=(0,0,255), thickness=25)
cv2.line(img=img_fix, pt1=pt1, pt2=pt3, color=(0,0,255), thickness=25)
cv2.line(img=img_fix, pt1=pt2, pt2=pt3, color=(0,0,255), thickness=25)

plt.imshow(img_fix)
````

````python
# Alternatively, using the polygon drawing method:
vertices = np.array([ [467,500], [267,900], [667,900] ], dtype = np.int32)
vertices.shape
````

````python
vertices
````

````python
points = vertices.reshape((-1,1,2)) # Cv2 requires the array to be in this 3-dim shape
points.shape
````

````python
points
````

````python
cv2.polylines(img_fix, [points], isClosed=True, color=(0, 0, 255), thickness=25)
plt.imshow(img_fix) 
````

```python
triangle_cnt = np.array( [pt1, pt2, pt3] )

cv2.drawContours(image, [triangle_cnt], 0, (0,255,0), -1)
```
Where:
cv2.drawContours(image, contours, contourIdx, color[, thickness[, lineType[, hierarchy[, maxLevel[, offset]]]]]) -> image
@param contourIdx Parameter indicating a contour to draw. If it is negative, all the contours are drawn.

````python
triangle_cnt = np.array( [pt1, pt2, pt3] )

cv2.drawContours(img_fix, [triangle_cnt], 0, (0,0,255), -1) #-1 is used to fill these contours

plt.imshow(img_fix)
````
